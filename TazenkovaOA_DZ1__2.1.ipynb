```python
X = df['Fat(g)']
Y = df['Calories']

MX = X.mean()
MY = Y.mean()

cov = ((X - MX) * (Y - MY)).mean()
cov
```

```python
cov = (X * Y).mean() - MX * MY
cov
```

```python
cov = np.cov(X, Y)
cov
```
